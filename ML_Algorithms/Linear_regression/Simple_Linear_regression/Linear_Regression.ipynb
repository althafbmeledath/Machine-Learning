Simple Linear Regression(Sklearn)

```python
#import the required packages
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
import seaborn as sns
import numpy as np
```

```python
#read the dataframe
df=pd.read_csv('placement.csv')
```

```python
#plot the data
plt.title('Linear Regression')
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
plt.show()
```

```python
#train test split
X=df.iloc[:,0:1].values
y=df.iloc[:,-1].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
#import and train the model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_test[0]
```

```python
lr.predict(X_test[0].reshape(1,1))
```

```python
#draw the line found by the algorithm
plt.title('Linear Regression')
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.plot(X_train,lr.predict(X_train),c='red')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
plt.show()
```

My Linear Regression

```python
class mylr:
    def fit(self,X_train,y_train):
        num=0
        den=0
        self.m=None
        self.b=None
        for i in range(X_train.shape[0]):
            num=num+(y_train[i]-y_train.mean())*(X_train[i]-X_train.mean())
            den=den+(X_train[i]-X_train.mean())**2
        self.m=num/den
        self.b=y_train.mean()-self.m*X_train.mean()
        print("Training completed.....")
        print(f'The slope is {m[0]}')
        print(f'The intercept is {b}')    
    def predict(self,X_test):
        return self.m*X_test+self.b
```

```python
lr1=mylr()
lr1.fit(X_train,y_train)
```

```python
lr1.predict(X_test)
```

```python
#draw the line found by my algorithm
plt.title('Linear Regression')
sns.scatterplot(x=df['cgpa'],y=df['package'])
plt.plot(X_train,lr1.predict(X_train),c='red')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
plt.show()
```
